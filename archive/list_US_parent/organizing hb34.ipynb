```python
import pandas as pd
import gspread
from gspread_dataframe import set_with_dataframe
from oauth2client.service_account import ServiceAccountCredentials

# Auth
scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
creds = ServiceAccountCredentials.from_json_keyfile_name('../gspread_creds.json', scope)
client = gspread.authorize(creds)
```

```python
# Open the sheet
worksheet = client.open("Handbuch 1934").worksheet("validated")

# Get all records as list of dicts
records = worksheet.get_all_records()

# Convert to DataFrame
hb1934 = pd.DataFrame(records)
```

```python
hb1934_w_us_firm = hb1934[hb1934['US parent']!='']
hb1934_w_us_firm
```

```python
hb1934_w_us_firm_to_csv = hb1934_w_us_firm.rename(columns={'Master US firm name': 'US Company'})
hb1934_w_us_firm_to_csv.to_csv('output/hb34.csv', index=False)
```
